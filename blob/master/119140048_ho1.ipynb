```python
import matplotlib.pyplot as plt
import numpy as np
import os
import wave
```

```python
file_path = os.path.join(os.getcwd(), 'data', 'SUPREME BEAM.wav')

#check
if not os.path.exists(file_path):
    print ('Not Found')
    exit()
```

```python
with wave.open(file_path, 'r') as wav_file:
    n_channels = wav_file.getnchannels()
    frames = wav_file.readframes(-1)
    sound_data = np.frombuffer(frames,dtype=np.int16)
    frame_rate = wav_file.getframerate()
    n_samples = wav_file.getnframes()

print(f"Jumlah channel: {n_channels}")
print(f"Frekuensi Sampling: {frame_rate} Hz")
print(f"Jumlah sampel: {n_samples}")
```
